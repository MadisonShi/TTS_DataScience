# How To Install R on a Mac

## In this documentation we will cover how to 

Install Anaconda  Install R  Install R studio 


### Install Anaconda

#### Installation
1. Go to https://www.anaconda.com/download/ and install the latest version of Anaconda available for MacOS
2. Double-click the downloaded file and run through the installer instructions


#### What is Anaconda?

Anaconda is a Python distribution platform, allowing the user to use the "conda install" command to access thousands of open-source Conda, Python, R, etc packages. Furthermore, Anaconda comes with the Anaconda Navigator which helps manage the packages and IDEs like Jupyter Notebook and R Studio. 

#### What is an IDE?

An integrated develop environment (IDE) is a software application that allows users to create code as well as compile and interpret the code in one software application.

#### Why use Anaconda?

Anaconda is useful because it allows the user to install a variety of data science related packages with a simple command. The user does not have to download each individual pieces of software needed.

### Install R 

#### Installation
1. In command prompt (assuming Anaconda is installed on the user's disk), use "conda install r-essentials --yes" command to install R

#### What is R 

R is a programming language and software environment designed for statistical programming and graphics. 

#### Who uses are R?

Statisticians and data scientists, as well as any academic or professional that works with data, uses R. 

#### How is R different from python?

R primarily used by academics for research purposes. R is great at exploratory data analysis. Python is largely used for data analysis and started as a way to teach students the basics of programming.

### Install R Studio 

#### Installation
1. Install from the Anaconda Navigator or write following command in terminal: "conda install -c r rstudio --yes"


#### What is R Studio?

R Studio is an integrated development environment for R, a programming language for statistical computing and graphics

#### What is a script?

A script is a computer language with a series of commands within a file capable of being executed without being compiled. 

#### What is a notebook?

A notebook is software that allows the user to display code, text, and images in a readable manner. It allows users to easily share their projects and findings with others.




#### What is the difference between R Notebook - R Markdown - Jupyter Notebook*** ( MUST ANSWER )

R Notebook and R Markdown are both part of the IDE R Studio. The primary difference is in their execution of code. R Notebooks are similar to Jupyter Notebooks and allow the user to execute code cell by cell (interactively) whereas R Markdown will execute code in batches. R Notebooks also allow users to use SQL, C++, R, Bash, and Python language.

Jupyter Notebook is a separate software that allows the users to create notebooks in a variety of styles and programming languages.



##### Tips: For any documentation or usage of the function in R Studio, just type the name of the function and then press cntrl+space to get the auto completion window.You can use ? before any function name to view the official documentation
