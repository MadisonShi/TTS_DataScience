```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
%matplotlib inline
import pandas as pd
```

```python
data = pd.read_csv('company_sales_data.csv')
data.head()
```

```python
#Exercise 1
plt.plot(data.month_number, data.total_profit)

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit Per Month')
```

```python
#Exercise 2
plt.plot(data.month_number, data.total_profit, ':ro', linewidth=3, markersize=5,label ='Monthly Profit')

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit Per Month')

plt.legend(loc='lower right')
```

```python
#Exercise 3
plt.plot(data.month_number, data.facecream)
plt.plot(data.month_number, data.facewash)
plt.plot(data.month_number, data.toothpaste)
plt.plot(data.month_number, data.bathingsoap)
plt.plot(data.month_number, data.shampoo)
plt.plot(data.month_number, data.moisturizer)

#for i in range(1,7):
#    plt.plot(data.month_number, sales[sales.columns[i]])
```

```python
#Exercise 4
plt.scatter(x=data.month_number, y=data.toothpaste)
plt.grid(linestyle='-')
plt.xticks(range(13))
```

```python
#Exercise 5

barWidth = 0.4

index = np.arange(1,13)
plt.xticks(index)

plt.bar(index,data.facecream, width=barWidth)
plt.bar(index+barWidth,data.facewash, width=barWidth)
```
