These data consist of Human Development Index and Corruption Perception Index scores for several countries.

Create a scatter plot with CPI on the x axis and HDI on the y axis.

Color the points blue.

Map the color of the the points to Region.

Make the points bigger by setting size to 2

Map the size of the points to HDI.Rank


```r
library(readr) 
data <- read_csv("EconomistData.csv")
library(tidyverse)
```

```r
#Create a scatter plot with CPI on the x axis and HDI on the y axis.
ggplot(data, aes(x=CPI,y=HDI)) +geom_point()
```

```r
#Color the points blue.
ggplot(data, aes(x=CPI,y=HDI)) +geom_point(color="blue")
```

```r
#Map the color of the the points to Region.
ggplot(data, aes(x=CPI,y=HDI, color=Region)) +geom_point()
```

```r
#Make the points bigger by setting size to 2
ggplot(data, aes(x=CPI,y=HDI, color=Region, size=2)) +geom_point()
```

```r
#Map the size of the points to HDI.Rank
ggplot(data, aes(x=CPI,y=HDI, color=Region, size=HDI.Rank)) +geom_point()
```

```r
#Overlay a smoothing line on top of the scatter plot using geom_smooth.
ggplot(data, aes(x=CPI,y=HDI)) +geom_point() + geom_smooth()
```

```r
#Overlay a smoothing line on top of the scatter plot using geom_smooth, but use a linear model for the predictions. Hint: see ?stat_smooth.
ggplot(data, aes(x=CPI,y=HDI)) +geom_point() + geom_smooth(method="lm")
```

```r
#Overlay a smoothing line on top of the scatter plot using geom_line. Hint: change the statistical transformation.
ggplot(data, aes(x=CPI,y=HDI)) +geom_point() + geom_line(stat = "smooth", method = "loess")
```

```r
#BONUS: Overlay a smoothing line on top of the scatter plot using the default loess method, but make it less smooth. Hint: see ?loess.

ggplot(data, aes(x = CPI, y = HDI)) + geom_point() + geom_smooth(span=0.20, method = "loess")
```

### Exercise III
Create a scatter plot with CPI on the x axis and HDI on the y axis. Color the points to indicate region.

Modify the x, y, and color scales so that they have more easily-understood names (e.g., spell out “Human development Index” instead of “HDI”).

Modify the color scale to use specific values of your choosing. Hint: see ?scale_color_manual.

```r
#Modify the x, y, and color scales so that they have more easily-understood names (e.g., spell out “Human development Index” instead of “HDI”).
ggplot(data, aes(x=CPI,y=HDI, color=Region)) +geom_point() + labs(y="Human development index", x="Corruption perception index") + scale_color_hue(labels=c("N. and S. Americas", "Asia Pacific", "Western Europe", "Eastern and Central Asia", "Middle East and Northern Africa", "Sub-Saharan Africa"))
```

```r
#Modify the color scale to use specific values of your choosing. Hint: see ?scale_color_manual.

ggplot(data, aes(x=CPI,y=HDI, color=Region)) +geom_point() + labs(y="Human development index", x="Corruption perception index") + scale_color_manual(labels=c("N. and S. Americas", "Asia Pacific", "Western Europe", "Eastern and Central Asia", "Middle East and Northern Africa", "Sub-Saharan Africa"), values=c("red", "orange", "yellow","green","blue", "purple"))
```
